```python
#Importing Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading File Walmart
walmart=pd.read_csv("walmart.csv")
walmart.head(11)
```

```python
print(walmart.shape)
walmart.info()
```

```python
# checking Null values in each column
walmart.isnull().sum()
```

```python
#Duplicate values
print(walmart.duplicated().sum())
walmart.drop_duplicates(inplace=True)
```

```python
walmart.shape
```

```python
# Converting str to float
walmart['unit_price']=walmart['unit_price'].astype(str).str.replace('$','').astype(float)
```

```python
walmart.head(11)
```

```python
# checking data distribution
sns.displot(walmart['unit_price'],kde=True)
```

```python
# checking data distribution
sns.displot(walmart['quantity'],kde=True)
```

```python
# Data Imputation with Mean and Median 
walmart['unit_price']=walmart['unit_price'].fillna(walmart['unit_price'].mean())
walmart['quantity']=walmart['quantity'].fillna(walmart['quantity'].median())
```

```python
walmart.isnull().sum()
```

```python
walmart['Sales']=walmart['unit_price']*walmart['quantity']
```

```python
walmart.head(11)
```

```python
# checking any duplicate or null values remaining
print(walmart.duplicated().sum())
print(walmart.isnull().sum().sum())
print(walmart.shape)
```

```python
# saving data file
walmart.to_csv("walmart.csv", index=False)
```
